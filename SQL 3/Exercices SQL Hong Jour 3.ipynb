##  Commandes D√©taill√©es
üëâ Obtenir le nom de contact du client et le pr√©nom de l'employ√© de toutes les commandes effectu√©es.

````python
# Query the sqlite_master table to get the schema of all tables
query = '''
SELECT employees.FirstName AS Prenom_Employ√©, customers.ContactName AS Nom_Client, customers.ContactTitle As Titre_Client 
FROM customers
JOIN orders ON orders.CustomerID = customers.CustomerId
JOIN employees ON employees.EmployeeID = orders.EmployeeID

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````

### D√©pense par Client

üëâ Listez le montant total d√©pens√© par client - dans l'ordre croissant. üëå

‚ÑπÔ∏è Nous ne prendrons pas en compte les frais de transport ici.

````python
# Query the sqlite_master table to get the schema of all tables
query = '''
SELECT orders.OrderID, customers.customerID, customers.ContactName AS Nom_Client, SUM(COALESCE(orderdetails.unitprice, 0) * COALESCE(orderdetails.quantity, 1)) AS Depense_Client
FROM customers
LEFT JOIN orders ON orders.CustomerID = customers.CustomerId
LEFT JOIN orderdetails ON orderdetails.orderID = orders.orderID
GROUP BY customers.CustomerID
ORDER BY Depense_Client ASC

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````

### Qui vend le plus ?

D√©terminer qui est le meilleur employ√© ! Par "meilleur employ√©", nous entendons celui qui vend le plus. üëë<br>
Nous attendons que la requ√™te retourne une valeur comme : `'FirstName', 'LastName', 6000 (la somme de tous les achats)`.

````python
# Query the sqlite_master table to get the schema of all tables
query = '''
SELECT orders.OrderID, employees.FirstName AS Prenom_Employ√©, employees.LastName AS Nom_Employ√©, sum(orderdetails.UnitPrice * orderdetails.Quantity) AS Somme_tous_achats
FROM employees
LEFT JOIN orders ON orders.EmployeeID = employees.EmployeeID
LEFT JOIN orderdetails ON orderdetails.orderID = orders.orderID
GROUP BY employees.EmployeeID
ORDER BY Somme_tous_achats DESC
LIMIT 1

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````

### Qui n'ach√®te rien ?

Rediger la requ√™te SQL adapt√©e pour obtenir le nombre de commandes effectu√©es par chaque client. Comme vous pouvez le deviner avec le titre, vous devriez √©galement afficher les clients sans commandes.

````python
# Query the sqlite_master table to get the schema of all tables
query = '''

SELECT customers.CustomerID, customers.ContactName, Count(orders.OrderId) AS Total_order
FROM Customers 
LEFT JOIN Orders ON Customers.CustomerID = Orders.CustomerID
LEFT JOIN OrderDetails ON Orders.OrderID = OrderDetails.OrderID
LEFT JOIN Employees ON Employees.EmployeeID = orders.employeeid
GROUP BY Customers.CustomerID
ORDER BY Total_order ASC

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````

## Donn√©es

## Sp√©cifications

**Indice :** Il y a deux sommes √† calculer ici : le montant total pour une commande donn√©e et le montant cumulatif des commandes. Laquelle est une window function ?

## Points cl√©s d'apprentissage

### Classement des commandes par client

- Classer les commandes de chaque client selon la date de commande :
- Pour chaque client, les commandes doivent √™tre class√©es dans l'ordre chronologique.
- Les informations (`OrderID`, `CustomerID`, `OrderDate`, `OrderRank`) doivent √™tre renvoy√©s.

````python
# Query the sqlite_master table to get the schema of all tables
query = '''

SELECT orders.orderID, orders.customerID, orders.OrderDate,
RANK() OVER (PARTITION BY orders.customerId ORDER BY orders.orderdate ASC) AS order_rank
FROM orders

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````

### Montant cumulatif des commandes par client

- Calculer le montant cumulatif (en USD) des commandes de chaque client selon la date de commande :
- Pour chaque client, les commandes doivent √™tre class√©es dans l'ordre chronologique.
- Les informations (`OrderID`, `CustomerID`, `OrderDate`, `OrderCumulativeAmount`) doivent √™tre renvoy√©s.

````python
# Query the sqlite_master table to get the schema of all tables
query = '''

SELECT orders.orderid, customers.customerid, orders.orderdate, sum(orderdetails.unitprice * orderdetails.quantity) as somme_ventes,

SUM(orderdetails.UnitPrice * orderdetails.Quantity) OVER(PARTITION BY orders.customerid ORDER BY orders.orderdate asc) as OrderCumulativeAmount

FROM customers

join orders on customers.customerid = orders.customerid

join orderdetails on orderdetails.orderid = orderdetails.orderid

GROUP BY orders.OrderID, orders.CustomerID, orders.OrderDate

ORDER BY orders.CustomerID, orders.OrderDate;

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df

````

### Moyenne par client

- Retourner le montant moyen d√©pens√© par commande pour chaque client, class√© par `CustomerID`.
- Les informations (`CustomerID`, `AverageOrderedAmount`) doivent √™tre renvoy√©s.

````python
# Query the sqlite_master table to get the schema of all tables
query = '''

WITH OrderTotal AS (SELECT o.CustomerID, od.OrderID, SUM(od.Quantity * od.UnitPrice) AS OrderAmount

FROM Orders o

JOIN OrderDetails od ON o.OrderID = od.OrderID

GROUP BY o.CustomerID, od.OrderID)

SELECT CustomerID, AVG(OrderAmount) AS AverageOrderedAmount

FROM OrderTotal

GROUP BY CustomerID

ORDER BY CustomerID

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````

### Moyenne g√©n√©rale

- Renvoyer le montant moyen d√©pens√© par commande.
- La requ√™te doit retourner un simple `float` (nombre decimal).

````python
# Query the sqlite_master table to get the schema of all tables
query = '''

WITH OrderTotal AS (SELECT o.CustomerID, od.OrderID, SUM(od.Quantity * od.UnitPrice) AS OrderAmount

FROM Orders AS o

JOIN OrderDetails od ON o.OrderID = od.OrderID

GROUP BY o.CustomerID, od.OrderID)

SELECT AVG(OrderAmount) AS AVG_Order_Amount

FROM OrderTotal

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````

### Meilleurs clients

Maintenant, trouvons les clients qui d√©pensent plus que la moyenne - c'est-√†-dire, leur montant moyen d√©pens√© par commande est sup√©rieur √† la moyenne g√©n√©rale d√©pens√©e par commande.

Voyez-vous que la partie principale a d√©j√† √©t√© trait√©e gr√¢ce aux 2 questions pr√©c√©dentes ? Utilisons nos requ√™tes pr√©c√©dentes avec √† la clause `WITH`.

- Renvoyer les clients qui ont une commande moyenne sup√©rieure √† la commande moyenne g√©n√©rale, tri√©s par montant moyen command√© d√©croissant.
- Les informations (`CustomerID`, `AverageOrderedAmount`) doivent √™tre renvoy√©s.

````python
# Query the sqlite_master table to get the schema of all tables
query = '''

WITH OrderTotal AS (SELECT o.CustomerID, od.OrderID, SUM(od.Quantity * od.UnitPrice) AS Order_Amount

FROM Orders o

JOIN OrderDetails od ON o.OrderID = od.OrderID

GROUP BY o.CustomerID, od.OrderID)

SELECT customerid, avg(Order_Amount) AS AverageOrderedAmount

From OrderTotal

Group by customerid

Having AverageOrderedAmount > 983.425

order by AverageOrderedAmount DESC

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````

### Produits les plus command√©s

- Renvoyer liste du produit le plus command√© (en termes de _montant d'argent_ et non de quantit√©) par client, bas√© sur le montant total command√© en **USD** et tri√© de mani√®re d√©croissante.
- Les informations (`CustomerID`, `ProductID`, `OrderedAmount`) doivent √™tre renvoy√©s.

````python
# Query the sqlite_master table to get the schema of all tables
query = '''

With OrderAmounts AS (select o.CustomerID, od.ProductID, SUM(od.Quantity * od.UnitPrice) AS OrderAmount

from Orders o

join OrderDetails od ON o.OrderID = od.OrderID

group by o.CustomerID, od.ProductID),

RankedProducts AS (select CustomerID, ProductID, OrderAmount, Rank() OVER (PARTITION BY CustomerID ORDER BY OrderAmount DESC) AS RowNumber

from OrderAmounts)

select CustomerID, ProductID, orderAmount

from RankedProducts

where RowNumber = 1

order by OrderAmount DESC

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````

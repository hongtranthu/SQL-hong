##  Commandes Détaillées
👉 Obtenir le nom de contact du client et le prénom de l'employé de toutes les commandes effectuées.

````python
# Query the sqlite_master table to get the schema of all tables
query = '''
SELECT employees.FirstName AS Prenom_Employé, customers.ContactName AS Nom_Client, customers.ContactTitle As Titre_Client 
FROM customers
JOIN orders ON orders.CustomerID = customers.CustomerId
JOIN employees ON employees.EmployeeID = orders.EmployeeID

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````

### Dépense par Client

👉 Listez le montant total dépensé par client - dans l'ordre croissant. 👌

ℹ️ Nous ne prendrons pas en compte les frais de transport ici.

````python
# Query the sqlite_master table to get the schema of all tables
query = '''
SELECT orders.OrderID, customers.customerID, customers.ContactName AS Nom_Client, SUM(COALESCE(orderdetails.unitprice, 0) * COALESCE(orderdetails.quantity, 1)) AS Depense_Client
FROM customers
LEFT JOIN orders ON orders.CustomerID = customers.CustomerId
LEFT JOIN orderdetails ON orderdetails.orderID = orders.orderID
GROUP BY customers.CustomerID
ORDER BY Depense_Client ASC

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````

### Qui vend le plus ?

Déterminer qui est le meilleur employé ! Par "meilleur employé", nous entendons celui qui vend le plus. 👑<br>
Nous attendons que la requête retourne une valeur comme : `'FirstName', 'LastName', 6000 (la somme de tous les achats)`.

````python
# Query the sqlite_master table to get the schema of all tables
query = '''
SELECT orders.OrderID, employees.FirstName AS Prenom_Employé, employees.LastName AS Nom_Employé, sum(orderdetails.UnitPrice * orderdetails.Quantity) AS Somme_tous_achats
FROM employees
LEFT JOIN orders ON orders.EmployeeID = employees.EmployeeID
LEFT JOIN orderdetails ON orderdetails.orderID = orders.orderID
GROUP BY employees.EmployeeID
ORDER BY Somme_tous_achats DESC
LIMIT 1

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````

````python
# Query the sqlite_master table to get the schema of all tables
query = '''

SELECT customers.CustomerID, customers.ContactName, Count(orders.OrderId) AS Total_order
FROM Customers 
LEFT JOIN Orders ON Customers.CustomerID = Orders.CustomerID
LEFT JOIN OrderDetails ON Orders.OrderID = OrderDetails.OrderID
LEFT JOIN Employees ON Employees.EmployeeID = orders.employeeid
GROUP BY Customers.CustomerID
ORDER BY Total_order ASC

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````

## Contexte & Objectifs

L'objectif de ce défi est d'utiliser des [Window functions](https://mode.com/sql-tutorial/sql-window-functions/) comme `RANK` ou `SUM` avec le mot-clé `OVER`.

La clause `OVER` peut être utilisée avec :

## Données

## Spécifications

### Classement des commandes par client

### Montant cumulatif des commandes par client

- Calculer le montant cumulatif (en USD) des commandes de chaque client selon la date de commande :
- Pour chaque client, les commandes doivent être classées dans l'ordre chronologique.

**Indice :** Il y a deux sommes à calculer ici : le montant total pour une commande donnée et le montant cumulatif des commandes. Laquelle est une window function ?

## Points clés d'apprentissage

- Les fonctions de fenêtrage vous permettent de calculer des données à partir de différentes lignes **sans agréger les lignes**.
- Si vous souhaitez agréger les lignes, utilisez `GROUP BY`, si ce n'est pas le cas, utilisez `OVER` avec `PARTITION BY` et `ORDER BY` à la place.


### Classement des commandes par client

- Classer les commandes de chaque client selon la date de commande :
- Pour chaque client, les commandes doivent être classées dans l'ordre chronologique.
- Les informations (`OrderID`, `CustomerID`, `OrderDate`, `OrderRank`) doivent être renvoyés.

````python
# Query the sqlite_master table to get the schema of all tables
query = '''

SELECT orders.orderID, orders.customerID, orders.OrderDate,
RANK() OVER (PARTITION BY orders.customerId ORDER BY orders.orderdate ASC) AS order_rank
FROM orders

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````

### Montant cumulatif des commandes par client

- Calculer le montant cumulatif (en USD) des commandes de chaque client selon la date de commande :
- Pour chaque client, les commandes doivent être classées dans l'ordre chronologique.
- Les informations (`OrderID`, `CustomerID`, `OrderDate`, `OrderCumulativeAmount`) doivent être renvoyés.

````python
# Query the sqlite_master table to get the schema of all tables
query = '''

SELECT orders.orderid, customers.customerid, orders.orderdate, sum(orderdetails.unitprice * orderdetails.quantity) as somme_ventes,

SUM(orderdetails.UnitPrice * orderdetails.Quantity) OVER(PARTITION BY orders.customerid ORDER BY orders.orderdate asc) as OrderCumulativeAmount

FROM customers

join orders on customers.customerid = orders.customerid

join orderdetails on orderdetails.orderid = orderdetails.orderid

GROUP BY orders.OrderID, orders.CustomerID, orders.OrderDate

ORDER BY orders.CustomerID, orders.OrderDate;

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df

````

### Moyenne par client

- Retourner le montant moyen dépensé par commande pour chaque client, classé par `CustomerID`.
- Les informations (`CustomerID`, `AverageOrderedAmount`) doivent être renvoyés.

````python
# Query the sqlite_master table to get the schema of all tables
query = '''

WITH OrderTotal AS (SELECT o.CustomerID, od.OrderID, SUM(od.Quantity * od.UnitPrice) AS OrderAmount

FROM Orders o

JOIN OrderDetails od ON o.OrderID = od.OrderID

GROUP BY o.CustomerID, od.OrderID)

SELECT CustomerID, AVG(OrderAmount) AS AverageOrderedAmount

FROM OrderTotal

GROUP BY CustomerID

ORDER BY CustomerID

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````

### Moyenne générale

- Renvoyer le montant moyen dépensé par commande.
- La requête doit retourner un simple `float` (nombre decimal).

````python
# Query the sqlite_master table to get the schema of all tables
query = '''

WITH OrderTotal AS (SELECT o.CustomerID, od.OrderID, SUM(od.Quantity * od.UnitPrice) AS OrderAmount

FROM Orders AS o

JOIN OrderDetails od ON o.OrderID = od.OrderID

GROUP BY o.CustomerID, od.OrderID)

SELECT AVG(OrderAmount) AS AVG_Order_Amount

FROM OrderTotal

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````

### Meilleurs clients

Maintenant, trouvons les clients qui dépensent plus que la moyenne - c'est-à-dire, leur montant moyen dépensé par commande est supérieur à la moyenne générale dépensée par commande.

Voyez-vous que la partie principale a déjà été traitée grâce aux 2 questions précédentes ? Utilisons nos requêtes précédentes avec à la clause `WITH`.

- Renvoyer les clients qui ont une commande moyenne supérieure à la commande moyenne générale, triés par montant moyen commandé décroissant.
- Les informations (`CustomerID`, `AverageOrderedAmount`) doivent être renvoyés.

````python
# Query the sqlite_master table to get the schema of all tables
query = '''

WITH OrderTotal AS (SELECT o.CustomerID, od.OrderID, SUM(od.Quantity * od.UnitPrice) AS Order_Amount

FROM Orders o

JOIN OrderDetails od ON o.OrderID = od.OrderID

GROUP BY o.CustomerID, od.OrderID)

SELECT customerid, avg(Order_Amount) AS AverageOrderedAmount

From OrderTotal

Group by customerid

Having AverageOrderedAmount > 983.425

order by AverageOrderedAmount DESC

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````

### Produits les plus commandés

- Renvoyer liste du produit le plus commandé (en termes de _montant d'argent_ et non de quantité) par client, basé sur le montant total commandé en **USD** et trié de manière décroissante.
- Les informations (`CustomerID`, `ProductID`, `OrderedAmount`) doivent être renvoyés.

````python
# Query the sqlite_master table to get the schema of all tables
query = '''

With OrderAmounts AS (select o.CustomerID, od.ProductID, SUM(od.Quantity * od.UnitPrice) AS OrderAmount

from Orders o

join OrderDetails od ON o.OrderID = od.OrderID

group by o.CustomerID, od.ProductID),

RankedProducts AS (select CustomerID, ProductID, OrderAmount, Rank() OVER (PARTITION BY CustomerID ORDER BY OrderAmount DESC) AS RowNumber

from OrderAmounts)

select CustomerID, ProductID, orderAmount

from RankedProducts

where RowNumber = 1

order by OrderAmount DESC

'''

# Create a pandas DataFrame from the SQL query
df = pd.read_sql_query(query, conn)

# Display the DataFrame
df
````
